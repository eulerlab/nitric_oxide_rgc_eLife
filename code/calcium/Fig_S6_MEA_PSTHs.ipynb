```python
def compute_cumsum(traces, window_size):
    
    # Number of windows (800 frames / 10-frame window)
    num_windows = traces.shape[1] // window_size

    # Initialize array to store cumulative firing rate for each neuron
    cumulative_firing_rates = np.zeros((traces.shape[0], num_windows))

    # Loop over each neuron
    for neuron_idx in range(traces.shape[0]):
        # Sum the PSTH data for each 10-frame window
        window_sums = np.array([np.sum(traces[neuron_idx, i:i+window_size]) 
                                for i in range(0, traces.shape[1], window_size)])

        # Compute cumulative firing rate for the neuron
        cumulative_firing_rates[neuron_idx, :] = np.cumsum(window_sums)

    # Compute the mean and standard deviation across neurons
    mean_cumulative_firing_rate = np.mean(cumulative_firing_rates, axis=0)
    std_cumulative_firing_rate = np.std(cumulative_firing_rates, axis=0)

    # Create x-axis values (each window corresponds to 10 frames)
    window_centers = np.arange(0, num_windows * window_size, window_size) + window_size // 2
    
    
    return mean_cumulative_firing_rate, std_cumulative_firing_rate, window_centers
```
